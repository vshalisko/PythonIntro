## Define a function to calculate admission score

### Subtask:
Create a function that takes a row of the DataFrame and the chosen department as input, and returns the calculated admission score based on the specified exams for that department.


**Reasoning**:
Define a function to calculate the admission score based on the department and relevant exam scores.



```python

# lista de departamentos
departments = ["Biotech","Chemistry","Engineering","Mathematics","Physics"]


def calculate_admission_score(row, department):
  """Calculates the admission score for a student based on their chosen department.

  Args:
    row: A pandas Series representing a student's data.
    department: A string representing the chosen department.

  Returns:
    The calculated admission score (mean of relevant exam scores).
  """
  if department == "Physics":
    exam_scores = [row["EPH"], row["EMA"], row["ESP"]]
  elif department == "Mathematics":
    exam_scores = [row["EMA"], row["ECS"], row["ESP"]]
  elif department == "Engineering":
    exam_scores = [row["EMA"], row["ECS"], row["EPH"], row["ESP"]]
  elif department == "Chemistry":
    exam_scores = [row["ECH"], row["EPH"], row["ESP"]]
  elif department == "Biotech":
    exam_scores = [row["EMA"], row["ECS"], row["ECH"], row["ESP"]]
  else:
    return None  # Handle cases with unknown departments

  return round(sum(exam_scores) / len(exam_scores), 2)
```

## Apply the function to each row and each option

### Subtask:
Iterate through the rows of the DataFrame and apply the function created in step 1 for each of the `OP1`, `OP2`, and `OP3` columns to calculate the respective scores.


**Reasoning**:
Iterate through the DataFrame and apply the `calculate_admission_score` function for each of the OP columns to calculate the scores and store them in new columns.



```python
df['OP1_score'] = df.apply(lambda row: calculate_admission_score(row, row['OP1']), axis=1)
df['OP2_score'] = df.apply(lambda row: calculate_admission_score(row, row['OP2']), axis=1)
df['OP3_score'] = df.apply(lambda row: calculate_admission_score(row, row['OP3']), axis=1)
display(df.head())
```

## Summary:

*   The admission score is the mean of relevant exam scores, with the required exams varying based on the department: Physics (EPH, EMA, ESP), Mathematics (EMA, ECS, ESP), Engineering (EMA, ECS, EPH, ESP), Chemistry (ECH, EPH, ESP), and Biotech (EMA, ECS, ECH, ESP).

```python
aspirants = pd.DataFrame(columns=["full_name", "department", "admission_score", "option"])
for index, row in df.iterrows():
    for option in ["OP1", "OP2", "OP3"]:
      full_name = row["firstname"] + " " + row["lastname"]
      aspirants.loc[-1] = [full_name, row[option], row[f"{option}_score"], option]
      aspirants.index += 1
      aspirants = aspirants.sort_index()

display(aspirants)
```

Read an integer N representing the maximum number of students per department.

```python
N = int(input("Enter the maximum number of students per department: "))
print(f"Maximum number of students per department (N): {N}")
```

```python
admitted_aspirants = pd.DataFrame(columns=["full_name", "department", "admission_score", "option"])

## aplicando OP1
for department in departments:
    department_aspirants = aspirants[(aspirants['department'] == department) & (aspirants['option'] == 'OP1')]
    department_aspirants_sorted = department_aspirants.sort_values(by=['admission_score'], ascending=[False])
    print(f"\nDepartment: {department}")
    print("Alumnos por admitir:")
    display(department_aspirants_sorted.head(N))
    if(admitted_aspirants.empty):
      admitted_aspirants = department_aspirants_sorted.head(N)
    else:
      admitted_aspirants = pd.concat([admitted_aspirants,department_aspirants_sorted.head(N)], ignore_index=True)

#display(admitted_aspirants)

list_of_admitted_aspirants = list(admitted_aspirants['full_name'])
#display(list_of_admitted_aspirants)
print("\nNúmero de alumnos admitidos OP1: ", str(len(list_of_admitted_aspirants)))

## generate updated list of aspirants by removing admitted
mask = ~aspirants['full_name'].isin(list_of_admitted_aspirants)
aspirants_cleaned = aspirants[mask]
#display(aspirants_cleaned)

## aplicando OP2
for department in departments:
    department_aspirants = aspirants_cleaned[(aspirants_cleaned['department'] == department) & (aspirants_cleaned['option'] == 'OP2')]
    department_aspirants_sorted = department_aspirants.sort_values(by=['admission_score'], ascending=[False])
    N_admitted = admitted_aspirants[admitted_aspirants['department'] == department].shape[0]
    N_pendientes = N - N_admitted
    print(f"\nDepartment: {department}")
    if N_pendientes <= 0:
      print("No more space in this department")
    else:
      print("Alumnos por admitir:")
      display(department_aspirants_sorted.head(N_pendientes))
      admitted_aspirants = pd.concat([admitted_aspirants,department_aspirants_sorted.head(N_pendientes)], ignore_index=True)


#display(admitted_aspirants)

list_of_admitted_aspirants = list(admitted_aspirants['full_name'])
#display(list_of_admitted_aspirants)
print("\nNúmero de alumnos admitidos OP1 + OP2: ", str(len(list_of_admitted_aspirants)))

## generate updated list of aspirants by removing admitted
mask = ~aspirants['full_name'].isin(list_of_admitted_aspirants)
aspirants_cleaned = aspirants[mask]
#display(aspirants_cleaned)



## aplicando OP3
for department in departments:
    department_aspirants = aspirants_cleaned[(aspirants_cleaned['department'] == department) & (aspirants_cleaned['option'] == 'OP3')]
    department_aspirants_sorted = department_aspirants.sort_values(by=['admission_score'], ascending=[False])
    N_admitted = admitted_aspirants[admitted_aspirants['department'] == department].shape[0]
    N_pendientes = N - N_admitted
    print(f"\nDepartment: {department}")
    if N_pendientes <= 0:
      print("No more space in this department")
    else:
      print("Alumnos por admitir:")
      display(department_aspirants_sorted.head(N_pendientes))
      admitted_aspirants = pd.concat([admitted_aspirants,department_aspirants_sorted.head(N_pendientes)], ignore_index=True)


#display(admitted_aspirants)

list_of_admitted_aspirants = list(admitted_aspirants['full_name'])
#display(list_of_admitted_aspirants)
print("\nNúmero de alumnos admitidos OP1 + OP2 + OP3: ", str(len(list_of_admitted_aspirants)))

## generate updated list of aspirants by removing admitted
mask = ~aspirants['full_name'].isin(list_of_admitted_aspirants)
aspirants_cleaned = aspirants[mask]
#display(aspirants_cleaned)
```
