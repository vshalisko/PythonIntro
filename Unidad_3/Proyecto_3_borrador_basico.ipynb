### Definir la función para el calculo de puntaje según el departamento de aspiración

##### Prompt **Gemini**

```
Definir una funcion que calcula el puntaje de admisión para aspirante
a un departamento de su elección. La función debe recibir como
entrada un DataFrame con las columnas que contienen el departamento
elegido por alumno (OP1, OP2, OP3) y las columnas con datos de examenes
(EPH, ECH, EMA, ECS y ESP ). Además la función debe recibir el parametro
de la columna que contiene la selección de departamento en este caso
(nombres OP1, OP2 u OP3). En cada ocasión la función debe calcular
el puntaje de admisión para la opción especificada y regresar
la columna con este valor. Para el cálculo del puntaje de admisión
dependen del departamento: Física: EPH, EMA, ESP Matemáticas:
EMA, ECS, ESP Ingeniería: EMA, ECS, EPH, ESP Química: ECH, EPH, ESP
Biotecnología: EMA, ECS, ECH, ESP El puntaje de admisión es el
promedio de las calificaciones correspondientes a los exámenes
relevantes, redondeado hasta dos digitos decimales. La función
debe servir para su posterior uso con el metodo .apply() para
calcular las nuevas columnas que van a contener las preferencias
de entrada.
```

````python
# función para calcular valor de puntaje para admisión,
# formula según el departamento
def calcular_puntaje_admision(linea, opcion):
  departmento = linea[opcion]

  if departmento == 'Physics':
    puntajes = [linea['EPH'], linea['EMA'], linea['ESP']]
  elif departmento == 'Mathematics':
    puntajes = [linea['EMA'], linea['ECS'], linea['ESP']]
  elif departmento == 'Engineering':
    puntajes = [linea['EMA'], linea['ECS'], linea['EPH'], linea['ESP']]
  elif departmento == 'Chemistry':
    puntajes = [linea['ECH'], linea['EPH'], linea['ESP']]
  elif departmento == 'Biotech':
    puntajes = [linea['EMA'], linea['ECS'], linea['ECH'], linea['ESP']]
  else:
    return None  # Para caso cuando departamento no esta en la lista

  return round(sum(puntajes) / len(puntajes), 2)
````

### Calcular las columnas adicionales en DataFrame

##### Prompt **Gemini**

```
Aplicar la función calcular_puntaje_admision() al DataFrame df para calcular
los valores de las columnas OP1_score para la opcion OP1, OP2_score para
la opcion OP2, OP3_score para la opcion OP3. Adicionalmente agregar
la columna nombre_completo con datos
```

````python
df['nombre_completo'] = df['nombre'] + ' ' + df['apellido']

df['OP1_score'] = df.apply(lambda r: calcular_puntaje_admision(r, 'OP1'), axis=1)
df['OP2_score'] = df.apply(lambda r: calcular_puntaje_admision(r, 'OP2'), axis=1)
df['OP3_score'] = df.apply(lambda r: calcular_puntaje_admision(r, 'OP3'), axis=1)

display(df.head())
````

### Eligir los mejores aspirantes al departamento

##### Prompt **Gemini**

```
Ordenar los aspirantes a departamento elegido por su puntaje de admisión,
tomando solo los datos de la columna OP1 como criterio de preferencia de
departamento y OP1_score como puntaje. Elegir N aspirantes con puntaje mas alto y guardar aspirantes elegidos en una tabla con cuatro columnas:
nombre_completo, departamento, puntaje, opción (valor 'OP1').
Presentar el contenido de la tabla con display() e imprimir el mensaje
sobre el número de aspirantes elegido.
```

````python
# Filtrar el DataFrame para el departamento seleccionado y la opción OP1
aspirantes_departamento = df[(df['OP1'] == departamento_seleccionado)]

# Ordenar por OP1_score en orden descendente
aspirantes_ordenados = aspirantes_departamento.sort_values(by=['OP1_score'], ascending=[False])

# Seleccionar los N aspirantes con los puntajes más altos
aspirantes_elegidos = aspirantes_ordenados.head(N)

# Crear la tabla con las columnas especificadas
tabla_aspirantes_elegidos = pd.DataFrame({
    'nombre_completo': aspirantes_elegidos['nombre_completo'],
    'departmento': aspirantes_elegidos['OP1'],
    'puntaje': aspirantes_elegidos['OP1_score'],
    'opcion': 'OP1'
})

# Presentar el contenido de la tabla
display(tabla_aspirantes_elegidos)

# Imprimir el número de aspirantes elegidos
print(f"\nNúmero de aspirantes elegidos para {departamento_seleccionado} (OP1): {len(tabla_aspirantes_elegidos)}")
````
